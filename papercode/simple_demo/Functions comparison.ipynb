```python
import numpy as np
import matplotlib.pyplot as plt

def source_function(x, x_shift):
    return (6*(x+x_shift) - 2)**2 * np.sin(12*(x+x_shift) - 4)

def sinx(x):
    return np.sin(x*np.pi*2)
def cosx(x):
    return -np.sin(x*np.pi*2)
def func1(x):
    return (x-0.5) ** 2.0
def func2(x):
    return -(1.4 - 3.0 * x) * np.sin(18.0 * x)




x = np.linspace(0, 1, 1000)  # 生成 x 的取值范围

plt.figure(figsize=(8, 6))  # 设置图像大小

# 设置全局字体为Arial，并设置默认字体大小
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.size'] = 16  # 设置默认字体大小

# 绘制 x_shift=0 的函数图像
y0 = source_function(x, 0)
plt.plot(x, y0, color='black', label='x_shift=0')



# # 绘制 x_shift=0.01 的函数图像

# y01 = source_function(x, +0.01)
# plt.plot(x, y01, color='red', alpha=1, label='x_shift=+0.01')

# # 绘制 x_shift=0.05 的函数图像


# y05 = source_function(x, +0.05)
# plt.plot(x, y05, color='blue', alpha=1, label='x_shift=+0.05')

# # 绘制 x_shift=0.1 的函数图像

# y1 = source_function(x, +0.1)
# plt.plot(x, y1, color='green', alpha=1, label='x_shift=+0.1')

# # # 绘制 x_shift=0.5 的函数图像

# # # y5 = 5*func2(x)
# # y5 = source_function(x, -0.5)
# # plt.plot(x, y5, color='purple', alpha=0.5, label='x_shift=-0.5')



y01 = 5*sinx(x)
plt.plot(x, y01, color='red', alpha=1, label='sinx')


y05 = 5*func1(x)
# y05 = 5*cosx(x)
plt.plot(x, y05, color='blue', alpha=1, label='func1')



y1 = 5*func2(x)
plt.plot(x, y1, color='green', alpha=1, label='func2')




# 设置标题
plt.title('Functions comparison', fontsize=18)

# 设置图例字体大小
plt.legend(fontsize=14)

# 设置坐标轴标签字体大小
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# 设置 x 轴和 y 轴标签
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)

plt.legend()
plt.grid(True)
plt.show()
```
