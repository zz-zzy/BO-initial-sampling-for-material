```python
fontsize = 20

color_sets = ['blue', 'green']
color_1 = (147/255, 85/255, 150/255)
color_2 = (116/255, 160/255, 161/255)

label_x = ' Creep Rupture Life (log) '
label_y = 'Counts'

# 创建一个2x2的子图
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# 获取所有数据的范围
all_y_values = np.concatenate([y_ti, y_ni, y_fe, y_co])
y_min, y_max = all_y_values.min()-1, all_y_values.max() +1

# 绘制Ti的y值分布直方图
sns.histplot(y_ti, kde=False, color='blue', ax=axes[0, 0])
# axes[0, 0].set_title('Ti', fontsize=fontsize)
axes[0, 0].set_xlabel(label_x, fontsize=fontsize)
axes[0, 0].set_ylabel(label_y, fontsize=fontsize)
axes[0, 0].set_xlim(y_min, y_max)

# 绘制Ni的y值分布直方图
sns.histplot(y_ni, kde=False, color='green', ax=axes[0, 1])
# axes[0, 1].set_title('Ni', fontsize=fontsize)
axes[0, 1].set_xlabel(label_x, fontsize=fontsize)
axes[0, 1].set_ylabel(label_y, fontsize=fontsize)
axes[0, 1].set_xlim(y_min, y_max)

# 绘制Fe的y值分布直方图
sns.histplot(y_fe, kde=False, color=color_1, ax=axes[1, 0])
# axes[1, 0].set_title('Fe', fontsize=fontsize)
axes[1, 0].set_xlabel(label_x, fontsize=fontsize)
axes[1, 0].set_ylabel(label_y, fontsize=fontsize)
axes[1, 0].set_xlim(y_min, y_max)

# 绘制Co的y值分布直方图
sns.histplot(y_co, kde=False, color=color_2, ax=axes[1, 1])
# axes[1, 1].set_title('Co', fontsize=fontsize)
axes[1, 1].set_xlabel(label_x, fontsize=fontsize)
axes[1, 1].set_ylabel(label_y, fontsize=fontsize)
axes[1, 1].set_xlim(y_min, y_max)

# 设置所有子图的x轴和y轴刻度字号
for ax in axes.flat:
    ax.tick_params(axis='x', labelsize=15)
    ax.tick_params(axis='y', labelsize=15)
    for spine in ax.spines.values():
        spine.set_linewidth(3)  # Set frame width to 3

# 调整子图之间的间距
plt.tight_layout()

# 显示图像
plt.show()
```
